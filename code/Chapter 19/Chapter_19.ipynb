````python
class Car:
    speed = 0
    started = False

    def start(self):
        self.started = True
        print("Car started, let's ride!")

    def increase_speed(self, delta):
        if self.started:
            self.speed += delta
            print("Vrooooooom!")
        else:
            print("You need to start the car first!")

    def stop(self):
        self.speed = 0
        self.started = False
        print("Car stopped.")


````

````python
class Car:
    def __init__(self, started = False, speed = 0):
        self.started = started
        self.speed = speed

    def start(self):
        self.started = True
        print("Car started, let's ride!")

    def increase_speed(self, delta):
        if self.started:
            self.speed += delta
            print("Vrooooooom1")
        else:
            print("You need to start the car first1")

    def stop(self):
        self.speed = 0
        self.started = False
        print("Car stopped.")
````

````python
class Vehicle:
    def __init__(self, started = False, speed = 0):
        self.started = started
        self.speed = speed

    def start(self):
        self.started = True
        print("Vehicle started, let's go!")

    def increase_speed(self, delta):
        if self.started:
            self.speed += delta
            print("Vrooooooom from Vehicle!")
        else:
            print("You need to start the vehicle first!")

    def stop(self):
        self.speed = 0
        print("Vehicle stopped.")
````

````python
class Car(Vehicle):
    trunk_open = False
    def open_trunk(self):
        self.trunk_open = True
    def close_trunk(self):
        self.trunk_open = False
````

````python
class Motorcycle(Vehicle):
    def __init__(self, started = False, speed = 0, center_stand_out = False):
        self.center_stand_out = center_stand_out
        super().__init__(started=started, speed=speed)
````

````python
motc = Motorcycle(started=True,speed=67, center_stand_out=True)
dir(motc)
print(motc.speed, motc.started)
````

````python
class Motorcycle(Vehicle):
    # 1. Accept specific motorcycle stuff first
    # 2. Accept *args and **kwargs for everything else
    def __init__(self, center_stand_out=False, *args, **kwargs):

        # Handle the specific Motorcycle stuff
        self.center_stand_out = center_stand_out

        # 3. Pass everything else up to the parent
        super().__init__(*args, **kwargs)
````

````python
motc = Motorcycle(started=True,speed=6767, center_stand_out=True)
dir(motc)
print(motc.speed, motc.started)
````
