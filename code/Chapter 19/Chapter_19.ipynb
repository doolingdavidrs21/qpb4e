````python
def say_hi(name):
    if name == '':
        print("You didn't enter your name!")
    else:
        print("Hi there...")
        for letter in name:
            print(letter)
````

````python
class Car:
    speed = 0
    started = False

    def start(self):
        self.started = True
        print("Car started, let's ride!")

    def increase_speed(self, delta):
        if self.started:
            self.speed += delta
            print("Vrooooooom!")
        else:
            print("You need to start the car first!")

    def stop(self):
        self.speed = 0
        self.started = False
        print("Car stopped.")


````

````python
car = Car()
car.increase_speed(10)
````

````python
car.start()
````

````python
car.increase_speed(40)
````

````python
car1 = Car()
car2 = Car()
id(car1)
````

````python
id(car2)
````

We created two car objects here, car1 and car2, and used the built-in method id() to get their ids. Each object in Python has a unique identifier, so we just proved that we created two different objects from the same class. We can use them independently:

````python
class Car:
    def __init__(self, started = False, speed = 0):
        self.started = started
        self.speed = speed

    def start(self):
        self.started = True
        print("Car started, let's ride!")

    def increase_speed(self, delta):
        if self.started:
            self.speed += delta
            print("Vrooooooom1")
        else:
            print("You need to start the car first1")

    def stop(self):
        self.speed = 0
        self.started = False
        print("Car stopped.")
````

````python
c1 = Car()
c2 = Car(True)
c3 = Car(True, 50)
c4 = Car(started=True, speed=40)
````
