### Try This: Using dictionaries
Suppose that you're writing a program that works like a spreadsheet. How might you use a dictionary to store the contents of a sheet? Write some sample code to both store a value and retrieve a value in a particular cell. What might be some drawbacks to this approach?

You could use tuples of row, column values as keys to store the values in a dictionary. One drawback would be that the keys wouldn’t be sorted, so you’d have to manage that situation as you grabbed the keys/values to render as a spreadsheet.



````python
# 1. Create an empty dictionary to represent the sheet
spreadsheet = {}

# 2. STORE values using a tuple (row, col) as the key
# Let's pretend Row 0 is the header row
spreadsheet[(0, 0)] = "Name"    # Row 0, Col 0
spreadsheet[(0, 1)] = "Age"     # Row 0, Col 1
spreadsheet[(0, 2)] = "City"    # Row 0, Col 2

# Let's add some data in Row 1
spreadsheet[(1, 0)] = "David"
spreadsheet[(1, 1)] = 54
spreadsheet[(1, 2)] = "Rio Rancho"

# 3. RETRIEVE a value
# We just call the key exactly as we stored it
print(f"The value at Row 1, Col 2 is: {spreadsheet[(1, 2)]}")
````

````python
# Create your solution in this cell
word_count = {}
with open("moby_01_clean.txt", "r") as infile:
    for word in infile:
        word = word.strip().lower()
        word_count[word] = word_count.get(word, 0) + 1

word_count_sorted = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

# top five most common words
print("Top 5 most common words:")
for word, count in word_count_sorted[:5]:
    print(word, count)


# five least common words
print("\nFive least common words:")
for word, count in word_count_sorted[-5:]:
    print(word, count)


````
